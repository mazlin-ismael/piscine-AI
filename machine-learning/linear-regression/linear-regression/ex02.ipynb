```python
from sklearn.datasets import make_regression
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
X, y, coef = make_regression(n_samples=100,
                         n_features=1,
                         n_informative=1,
                         noise=10,
                         coef=True,
                         random_state=0,
                         bias=100.0)
plt.scatter(X, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("X (1 dimension) and y")
```

```python
model = linear_model.LinearRegression()
model.fit(X, y)
print(f"{model.coef_[0]} * x + {model.intercept_}")
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X), c="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("X (1 dimension) and y")
```

```python
y_pred = model.predict(X)
y_pred[:10]
```

```python
def compute_mse(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    return mse
```

```python
compute_mse(y, y_pred)
```

```python
X1, y1, coef1 = make_regression(n_samples=100,
                         n_features=1,
                         n_informative=1,
                         noise=50,
                         coef=True,
                         random_state=0,
                         bias=100.0)

model1 = linear_model.LinearRegression()
model1.fit(X1, y1)
print(f"{model1.coef_[0]} * x + {model1.intercept_}")

plt.scatter(X1, y1)
plt.plot(X1, model1.predict(X1), c="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("X (1 dimension) and y")
plt.show()
```
