```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
X = [[0],[0.1],[0.2], [1],[1.1],[1.2], [1.3]]
y = [0,0,0,1,1,1,0]
```

```python
def sigmoid1(x):
    return 1/(1+ np.exp(-(0.5*x + 3)))
```

```python
def sigmoid2(x):
    return 1/(1+ np.exp(-(5*x + 11)))
```

```python
log = LogisticRegression()
log.fit(X, y)
x_tests = np.arange(-20, 15, 0.1).reshape(-1, 1)
y_preds = log.predict_proba(x_tests)[:, 1]
```

```python
plt.plot(x_tests, y_preds)
plt.plot(x_tests, sigmoid1(x_tests))
plt.plot(x_tests, sigmoid2(x_tests))
#plt.plot(x_tests, np.repeat(0.5, x_tests.shape[0]))
plt.axhline(y=0.5, c="red")
plt.title("Sigmoids")
plt.xlabel("x")
plt.show()
```
