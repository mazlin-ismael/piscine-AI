```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("auto-mpg.csv", na_values=["?", ""])
df.dropna(inplace=True)
```

```python
df.drop(columns=["model year", "origin", "car name"], inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns="mpg"), df["mpg"], test_size=0.2, shuffle=False
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_train.shape)
print(X_train[:5])
print(y_train[:5])
print(X_test.shape)
print(X_test[:5])
print(y_test[:5])
```

```python
model = Sequential()
model.add(Dense(16, input_shape=(X_train.shape[1], ), activation="sigmoid"))
model.add(Dense(8, activation="sigmoid"))
model.add(Dense(1, activation="linear"))
```

```python
model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)
```

```python
model.fit(X_train, y_train, epochs=1000)
```

```python
y_pred = model.predict(X_test)
model.evaluate(X_test, y_test)
```
