```python
import numpy
import pandas
import matplotlib.pyplot as plt
```

```python
left_data = [5, 7, 11, 13, 17]
right_data = [0.1, 0.2, 0.4, 0.8, -1.6]
x_axis = [0.0, 1.0, 2.0, 3.0, 4.0]
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(x_axis, left_data, c="black")
ax2.plot(x_axis, right_data, c="red")

ax1.set_ylabel("Big")
ax2.set_ylabel("Small")
plt.title("Twin axes")
```
