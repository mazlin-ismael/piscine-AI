```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 4, 5, 6, 7])
y = np.array([2, 6, 3, 6, 3])
```

```python
plt.scatter(x, y, c="blue", linewidth=12, zorder=2)
plt.plot(x, y, linestyle="dashdot", c="red", linewidth=3, zorder=1)

plt.title("Display marker")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim(1, 8)
plt.ylim(1, 8)
```
