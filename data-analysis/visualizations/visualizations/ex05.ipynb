```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2, 3)
plt.subplots_adjust(wspace=0.5, hspace=0.5)

for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.title(f"Title {i}")
    plt.text(0.5, 0.5, f"(2, 3, {i})", ha="center", fontsize=18)
```
