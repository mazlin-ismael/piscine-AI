```python
import pandas as pd
import numpy as np
```

```python
business_dates = pd.bdate_range('2021-01-01', '2021-12-31')

#generate tickers
tickers = ['AAPL', 'FB', 'GE', 'AMZN', 'DAI']

#create indexs
index = pd.MultiIndex.from_product([business_dates, tickers], names=['Date', 'Ticker'])

# create DFs
market_data = pd.DataFrame(index=index,
                        data=np.random.randn(len(index), 1),
                        columns=['Prediction'])
market_data
```

```python
unstacked_df = market_data.unstack()
unstacked_df.head().index
```

```python
unstacked_df.plot(y="Prediction", use_index=True, title="Tickers Predictions YEAR 2021")
# unstacked.plot(title = 'Stocks 2021'
```
